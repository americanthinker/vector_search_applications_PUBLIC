```python
# from huggingface_hub import notebook_login
from google.colab import userdata
import json
import requests
import numpy as np
from concurrent.futures import ThreadPoolExecutor, as_completed
import pandas as pd
```

```python
hf_token = userdata.get('HF_TOKEN')
```

```python
API_URL = "https://ztpy6dk0pkhqhncd.us-east-1.aws.endpoints.huggingface.cloud"
headers = {"Authorization": f"Bearer {hf_token}", "Content-Type": "application/json",}

def query(payload: str="Just say hi!", temp: float=0) -> dict:
    json_body = {
        "inputs": f"[INST] <<SYS>> Your job is to talk like James Bond. Every reponse must sound like James Bond. <<SYS>> {payload} [/INST] ",
                "parameters": {"max_new_tokens":100, "top_p":0.9, "temperature":temp}
        }
    data = json.dumps(json_body)
    response = requests.request("POST", API_URL, headers=headers, data=data)
    try:
        return json.loads(response.content.decode("utf-8"))
    except:
        return response

data = query()
print(data[0]['generated_text'])
```
